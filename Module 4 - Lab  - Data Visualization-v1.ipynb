**Read the CSV File into a Pandas DataFrame**

Load the Stack Overflow survey data into a Pandas DataFrame:


```python
# Read the CSV file
df = pd.read_csv('survey-data.csv')

# Display the first few rows of the data
df.head()
```

**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

**Verify the Data in the SQLite Database**
Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

For example, you can group data by a specific column, like Age, to get the count of respondents in each age group:


```python
QUERY = """
SELECT Age, COUNT(*) as count
FROM main
GROUP BY Age
ORDER BY Age
"""
pd.read_sql_query(QUERY, conn)
```

## Hands-on Lab


### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
## Write your code here'
QUERY = "SELECT CompTotal FROM main"
df = pd.read_sql_query(QUERY, conn)
df['CompTotal'] = pd.to_numeric(df['CompTotal'], errors='coerce')

plt.figure(figsize=(10, 6))
plt.hist(df['CompTotal'], bins=20, edgecolor='k')
plt.xlabel('Total Compensation')
plt.ylabel('Count')
plt.title('Distribution of Total Compensation')
plt.grid(True)
plt.show()
```

**Box Plots**

Plot a box plot of Age.


```python
## Write your code here
QUERY = "SELECT Age FROM main"
df2 = pd.read_sql_query(QUERY, conn)
# Create a age mapping for age calculation
age_mapping = {                                              
    'Under 18 years old': 17,
    '18-24 years old': (18 + 24) / 2,
    '25-34 years old': (25 + 34) / 2,
    '35-44 years old': (35 + 44) / 2,
    '45-54 years old': (45 + 54) / 2,
    '55-64 years old': (55 + 64) / 2,
    '65 years or older': 70,
    'Prefer not to say': 30
}
df2['Age'] = df2['Age'].map(age_mapping)

plt.figure(figsize=(10, 6))
plt.boxplot(df2['Age'], patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.ylabel('Age')
plt.title('Box Plot of Age')
plt.grid(True)
plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
## Write your code here
QUERY = "SELECT WorkExp FROM main"
df2['WorkExp'] = pd.read_sql_query(QUERY, conn)

df2['WorkExp'] = pd.to_numeric(df2['WorkExp'], errors='coerce')
df2['WorkExp']  = df2['WorkExp'].fillna(0)

plt.figure(figsize=(10, 6))
plt.scatter(df2['Age'], df2['WorkExp'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('WorkExp')
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
## Write your code here

#Create TimeSearching data
QUERY = "SELECT TimeSearching FROM main"
df2['TimeSearching'] = pd.read_sql_query(QUERY, conn)
df2['TimeSearching'].fillna('UnKnown', inplace=True)

#Create Frunstration data
QUERY = "SELECT Frustration FROM main"
df_frustration = pd.read_sql_query(QUERY, conn)

# Function to split and expand the 'Frustration' column
def expand_frustration(df):
    expanded_data = []
    for index, row in df.iterrows():
        frustrations = str(row['Frustration']).split(';')
        for frustration in frustrations:
            expanded_data.append({'Frustration': frustration})
    return pd.DataFrame(expanded_data)

expanded_df = expand_frustration(df_frustration)
df2['Frustration'] = expanded_df['Frustration']
df2['Frustration'].fillna('UnKnown', inplace=True)



plt.figure(figsize=(10, 6))
plt.xticks(rotation=90) 
plt.scatter(df2['TimeSearching'], df2['Frustration'], s=df2['Age'], alpha=0.1)
plt.xlabel('TimeSearching')
plt.ylabel('Frustration')
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
## Write your code here
QUERY = "SELECT DatabaseWantToWorkWith FROM main"
df_database = pd.read_sql_query(QUERY, conn)

df_database['DatabaseWantToWorkWith'].dropna(inplace=True)
# Function to split and expand the column
def expand(df,col):
    expanded_data = []
    for index, row in df.iterrows():
        data = str(row[col]).split(';')
        for cols in data:
            expanded_data.append({col: data})
    expanded_data=filter(None,expanded_data)
    return pd.DataFrame(expanded_data)

expanded_df = expand(df_database, 'DatabaseWantToWorkWith' )
#database_df=expanded_df['DatabaseWantToWorkWith'].value_counts()
#print(database_df['DatabaseWantToWorkWith'].value_counts())
fig, ax = plt.subplots(figsize=(10, 6))
top_5_databases = expanded_df['DatabaseWantToWorkWith'].value_counts().head(5)
top_5_databases.plot(kind='pie', autopct='%1.1f%%', ax=ax)
ax.set_ylabel('')
```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
## Write your code here

QUERY = """
SELECT TimeAnswering, TimeSearching
FROM main
WHERE Age = '25-34 years old'
"""
df_time = pd.read_sql_query(QUERY, conn)


# Calculate medians
median_time_answering = df_time['TimeAnswering'].value_counts().median()
median_time_searching = df_time['TimeSearching'].value_counts().median()

# Create the stacked bar chart
plt.figure(figsize=(8, 6))
plt.bar(['Age Group 25-34'], [median_time_searching], label='TimeSearching')
plt.bar(['Age Group 25-34'], [median_time_answering], bottom=[median_time_searching], label='TimeAnswering')
plt.xlabel('Age Group')
plt.ylabel('Median Time (minutes)')
plt.title('Median TimeSearching and TimeAnswering for Age Group 25-34')
plt.legend()
plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
## Write your code here

QUERY = """
SELECT Age, CompTotal
FROM main
WHERE Age BETWEEN '45-54 years old' AND '55-64 years old'
"""
df_comptotal = pd.read_sql_query(QUERY, conn)
plt.figure(figsize=(12, 6))
df_comptotal.groupby('Age')['CompTotal'].median().plot(kind='line', marker='o')
plt.xlabel('Age')
plt.ylabel('Median CompTotal')
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
## Write your code here
QUERY = "SELECT MainBranch FROM main"
df_mainbranch = pd.read_sql_query(QUERY, conn)
df_mainbranch['MainBranch'].value_counts().plot(kind='barh')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.title('MainBranch Distribution')
plt.show()

```
